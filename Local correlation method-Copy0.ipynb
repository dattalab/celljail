```python
import numpy as np
import scipy.ndimage as ndimage
import scipy.signal as signal
import skimage
import os
import tiffile
from glob import glob
from itertools import product
from pylab import *
```

```python
# Specify where the images are

datadir = "/Volumes/Neurobio/DattaLab/Paul/screendataforkei/screenNofK/"
img_dir = os.path.join(datadir, 'day5_073014_16hour/6c_0.125_293normal/try1/*small*')
img_temp = glob(img_dir)
img_files = sorted(img_temp)[:220]
print img_files
```

```python
# Load in the images
Is = np.rollaxis(np.dstack([tiffile.imread(i) for i in img_files]), 2, 0)
tiffile.imsave('6c.tif', Is[:], compress=0)
Is = Is[:,::2,::2].astype('float64')
```

```python
figure(figsize=(10,10))
imshow(np.log(np.max(Is[:219],0)**2.0), cmap='jet')
```

```python
figure(figsize=(20,8))
gs = GridSpec(1,2)
subplot(gs[0])
title("Log Maximum across frames")
imshow(np.log(Is.max(0))); colorbar(); axis('tight')
subplot(gs[1])
title("Mean across frames")
imshow(Is.mean(0)); colorbar(); axis('tight')
```

```python
# Calculate the maximum across all frames
accum = np.log(Is.max(axis=0))
```

```python
figure(figsize=(12,7))
imshow(accum); colorbar()
```

```python
import skimage.segmentation, skimage.feature, skimage.morphology.watershed
```

```python

figure(figsize=(20,20))
I = accum.copy()
# I = ndimage.grey_opening(I,(3,3))
maxes = skimage.feature.peak_local_max(I, min_distance=1.5, threshold_abs=4.9, indices=False)

I_show = accum.copy()
# I_show[I_show<0] = 0
# I_show = np.sqrt(I_show)

I_mask = ndimage.grey_dilation(maxes,(1,1)).astype('float32')
I_mask[I_mask == True] = I_show.max()*1.01
# figure(figsize=(20,7))
imshow(I_show+I_mask, cmap='jet', clim=(4,7.5)); colorbar()
```

```python
frames, x, y = Is.shape
label_mask, num_cells = ndimage.label(I_mask.astype(bool))
print frames
print num_cells
```

```python
traces = np.zeros((frames, num_cells))
baselined_traces = np.zeros_like(traces)
normed_traces = np.zeros_like(traces)
splined_trace = np.zeros_like(traces)
imshow(label_mask)
```

```python
for cell in range(1,num_cells):
    traces[:,cell] = Is[:,label_mask==cell].mean(axis=1)
    baselined_traces[:,cell] = traces[:,cell] - traces[:30,cell].mean()
    normed_traces[:,cell] = traces[:,cell] / traces[:30,cell].mean()
    
print "Found %d unique cells" % traces.shape[1]
```

```python
splinedBackgroundTrace = signal.spline_filter(traces, lmbda=0.5)
plot(splinedBackgroundTrace[:,:])
```

```python
for cell in range(1,num_cells):
    splined_trace[:,cell] = traces[:,cell] / splinedBackgroundTrace[:,cell]
```

```python
figure(figsize=(15,3))
title = "_Normalized(divided by baseline)"
suptitle(title)
frame = 1.0/250.0
axhspan(normed_traces.min(), normed_traces.max(),frame*30, frame*50, alpha=0.25)
axhspan(normed_traces.min(), normed_traces.max(),frame*140, frame*160, alpha=0.25)
# plot(normed_traces[:-220,:]);
plot(normed_traces[:-5,:]);
ylim(0.5,2)
```
