```python
def slope(startPoint,endPoint):
    try:
        return (float(endPoint[1])-startPoint[1])/(float(endPoint[0])-startPoint[0])
    except ZeroDivisionError:
        # line is vertical
        return None
def yintercept(point, slope):
    if slope != None:
        return point[1]-point[0]*slope
    else:
        return None
def linepoint(slope, yintercept,x):
    return slope*x+yintercept
```

```python
baselinesArray=np.zeros_like(traces)
baselinesArray.shape
```

```python
for cell in range(1,num_cells):
    startPoint=[0,lowpassedTraces[0,cell]]
    endPoint=[frames,lowpassedTraces[lowpassedTraces.shape[0]-1,cell]]
    m = slope(startPoint,endPoint)
    b = yintercept(startPoint,m)
    baseline=[]
    for frame in range(1,frames+1):
        basepoint = linepoint(m,b,frame)
        baseline.append(basepoint)
    baselinesArray[:,cell]=baseline

    
for cell in range(1,num_cells):
    baselined_traces[:,cell] = (traces[:,cell]-baselinesArray[:,cell])
```
