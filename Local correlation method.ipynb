```python
import numpy as np
import scipy.ndimage as ndimage
import skimage
import os
import tiffile
from glob import glob
from joblib import Parallel, delayed
from itertools import product

def kern(windowsize=7,i=0,j=0):
    k = np.zeros((1,windowsize,windowsize))
    k[:,i,j]=-1
    k[:,windowsize/2,windowsize/2]=1
    return k

def go(k):
    a = ndimage.correlate(Is, k)
    b = a.copy()
    a[a<0] = 0
    b[b>0] = 0
    return np.sqrt(np.sum(a**2.0,0)) - np.sqrt(np.sum(b**2.0,0))
```

```python
# Specify where the images are
datadir = "/Volumes/Data/paul_imaging/"
img_dir = os.path.join(datadir, '4c/mix7b/*small*')
img_files = glob(img_dir)
```

```python
# Load in the images
Is = np.rollaxis(np.dstack([tiffile.imread(i) for i in img_files]), 2, 0)
Is = Is[:,::2,::2].astype('float64')
```

```python
figure(figsize=(10,10))
imshow(np.log(np.max(Is,0)**2.0), cmap='jet')
```

```python
windowsize=21
# WATCH OUT FOR THIS! ADD A COMMAND-LINE OPTION FOR THE NUMBER OF PROCESSORS USED
Isf = Parallel(n_jobs=24)\
                (delayed(go)\
                (kern(windowsize=windowsize, i=i,j=j)) for i,j in product(range(windowsize),range(windowsize)))
    
accum = np.zeros_like(Isf[0])
for I in Isf:
    accum += I
```
