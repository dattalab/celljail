```python
# Load in the images
Is = np.rollaxis(np.dstack([tiffile.imread(i) for i in img_files]), 2, 0)
tiffile.imsave('6c.tif', Is[:], compress=0)
Is = Is[:,::2,::2].astype('float64')
```

```python
# Calculate the maximum across all frames
accum = np.log(Is.max(axis=0))
```

```python
I_show = accum.copy()
imshow(I_show)
```

```python
#increase contrast
import ImageEnhance
I_contrast = I_show*2
I_contrast = I_contrast.astype("uint16")
imshow(I_contrast)
```

```python
#find regional maxima
regionalMax = pymorph.regmax((I_contrast).astype("uint16"))
pylab.imshow(pymorph.overlay(I_contrast, regionalMax))

#find seeds and cell number
seeds,numCells = ndimage.label(regionalMax)
print numCells
```

```python

#edge detection
T = mahotas.thresholding.otsu(I_contrast.astype("uint16"))
dist = ndimage.distance_transform_edt(I_contrast.astype("uint16") > T)
dist = dist.max() - dist
dist -= dist.min()
dist = dist/float(dist.ptp()) * 255
dist = dist.astype(np.uint8)
pylab.imshow(dist)
```

```python
I_mask = pymorph.cwatershed(dist, seeds)
pylab.imshow(I_mask)
```

```python
frames, x, y = Is.shape
label_mask, num_cells = ndimage.label(I_mask.astype(bool))
print frames
print num_cells
```

```python
traces = np.zeros((frames, num_cells))
baselined_traces = np.zeros_like(traces)
normed_traces = np.zeros_like(traces)
splined_trace = np.zeros_like(traces)
# imshow(label_mask)
```

```python
m = tm.mask_all_deviations(traces, 2.25)
bs = tm.baseline_splines(m, 5)

#raw traces
figure(figsize=(20,7))
_ = plt.plot(traces)

#masked raw traces
figure(figsize=(20,7))
_ = plt.plot(m)

#baselines
plt.figure()
figure(figsize=(20,7))
_ = plt.plot(bs)

#normed_traces
plt.figure()
figure(figsize=(20,7))
frame = 1.0/GRAPHSIZE;
normed_traces = (traces-bs)/bs
axhspan(normed_traces.min(), normed_traces.max(),frame*STIM_1_START, frame*STIM_1_END, alpha=0.25);
axhspan(normed_traces.min(), normed_traces.max(),frame*STIM_2_START, frame*STIM_2_END, alpha=0.25);
_ = plt.plot(normed_traces)
plt.figure()

#smoothed normed traces
figure(figsize=(20,7))
axhspan(normed_traces.min(), normed_traces.max(),frame*STIM_1_START, frame*STIM_1_END, alpha=0.25);
axhspan(normed_traces.min(), normed_traces.max(),frame*STIM_2_START, frame*STIM_2_END, alpha=0.25);
for cell in range(1,num_cells):
    _ = plot(tm.smooth(normed_traces[:,cell], window_len=11, window='flat'))





```
